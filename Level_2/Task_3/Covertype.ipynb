```python
import pandas as pd


df=pd.read_csv("covertype.csv")
df.head()
```

```python
print(df.shape)
df.info()
df.describe()
```

```python
df.nunique()
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import seaborn as sns

counts=df["Cover_Type"].value_counts()

plt.pie(counts,labels=counts.index,autopct="%1.1f%%", startangle=90)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_sample_weight #if classes are imbalanced

X=df.drop(columns="Cover_Type")
df["Cover_Type"]=df["Cover_Type"]-1
y=df["Cover_Type"]
train_X,test_X,train_y,test_y=train_test_split(X,y,train_size=0.8,random_state=42)
model=RandomForestClassifier(n_estimators=75)
sample_weights = compute_sample_weight(class_weight="balanced", y=train_y)
model.fit(train_X, train_y, sample_weight=sample_weights)
y_pred=model.predict(test_X)
```

```python
model
```

```python
print(f"this is our Model Accuracy : {accuracy_score(test_y,y_pred)}")
cm=confusion_matrix(test_y,y_pred)
disp=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=model.classes_)
disp.plot(cmap="Blues")
plt.show()
```

```python
importances=model.feature_importances_

feat_importances=pd.Series(importances,index=train_X.columns)
feat_importances.nlargest(20).plot(kind="bar")
plt.title("RandomForest Feature Importance")
plt.show()
```

```python
import xgboost as sgb
from xgboost import XGBClassifier


model = XGBClassifier( objective="multi:softmax",
    num_class=7,                 
    n_estimators=350,
    max_depth=10,
    gamma=0.05,
    learning_rate=0.4,
    random_state=42,
    eval_metric="mlogloss")



sample_weights = compute_sample_weight(class_weight="balanced", y=train_y)
model.fit(train_X, train_y, sample_weight=sample_weights)

y_pred=model.predict(test_X)
```

```python
print(f"this is our Model Accuracy : {accuracy_score(test_y,y_pred)}")
cm=confusion_matrix(test_y,y_pred)
disp=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=model.classes_)
disp.plot(cmap="Blues")
plt.show()
```

```python
# Get feature importance
importances = model.feature_importances_

# Convert to pandas Series
feat_importances = pd.Series(importances, index=train_X.columns)

# Plot top 20
feat_importances.nlargest(20).plot(kind='barh')
plt.title("XGBoost Feature Importance (gain-based by default)")
plt.show()
```
